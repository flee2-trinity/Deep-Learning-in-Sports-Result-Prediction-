```python
import csv
filename = '/content/gdrive/My Drive/2021 Fall/Thesis I/Records/AllTeams1819_1.txt'
with open(filename, newline='') as f:
  reader = csv.reader(f)
  p1content = list(reader)
print(len(p1content))
```

```python
from collections import defaultdict
groups = defaultdict(list)
for t in range(len(p1content)):
  s = p1content[t][0].split()
  if s[0] != '-' :
    groups[int(s[2])].append(float(s[0]))
gList = list(groups.values())
```

```python
from statistics import mean
y1ListM = list(map(lambda d: mean(d), gList))
xList = [i+5 for i in range(len(y1ListM))]
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot(xList, y1ListM, label = "18-19")
plt.legend()
plt.show()
```
