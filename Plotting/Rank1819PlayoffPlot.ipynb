```python
import csv
filename = '/content/gdrive/My Drive/2021 Fall/Thesis I/Records/Rank1819Playoff.txt'
with open(filename, newline='') as f:
  reader = csv.reader(f)
  p1content = list(reader)
```

```python
xList = [i for i in range(163)]
y1List = [[]for i in range(163)]
y2List = [[]for i in range(163)]
y3List = [[]for i in range(163)]
y4List = [[]for i in range(163)]
```

```python

for t in range(len(p1content)):
  s = p1content[t][0].split()
  if s[0] != '-' :
    i = int(s[0])
    y1List[i].append(float(s[1]))
    y2List[i].append(float(s[2]))
    y3List[i].append(float(s[3]))
    y4List[i].append(float(s[4]))
```

```python
from statistics import mean
y1ListM = list(map(lambda d: mean(d), y1List))
y2ListM = list(map(lambda d: mean(d), y2List))
y3ListM = list(map(lambda d: mean(d), y3List))
y4ListM = list(map(lambda d: mean(d), y4List))
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot(xList, y1ListM, label = "1819Power Ranking West")
plt.plot(xList, y2ListM, label = "Power Ranking East")
plt.plot(xList, y3ListM, label = "Total Ranking West")
plt.plot(xList, y4ListM, label = "Total Ranking East")
plt.legend()
plt.show()
```

```python
from scipy import stats

all_points = stats.ttest_ind(y1ListM, y2ListM)
print(all_points)
```
