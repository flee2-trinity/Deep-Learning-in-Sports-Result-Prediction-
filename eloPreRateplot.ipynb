```python
import csv
filename = '/content/gdrive/My Drive/2021 Fall/Thesis I/Records/eloPreRate.txt'
with open(filename, newline='') as f:
  reader = csv.reader(f)
  p1content = list(reader)
```

```python
xList = [i for i in range(2008, 2019)]
#season, (1 or 0)
```

```python
y1List = [[] for _ in range(2023)]
for line in p1content:
  a, b = line[0].split()
  season, r = int(a), int(b)
  y1List[season].append(r)
```

```python
y1ListR = []
for x in xList:
  print(x)
  y1ListR.append(sum(y1List[x])/len(y1List[x]))
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot(xList, y1ListR, label = "Elo Prediction 2008 - 2018")
plt.xlabel('Year') 
plt.ylabel('Prediction Rate') 
plt.rcParams["figure.figsize"] = (10,7)
plt.legend()
plt.show()
```
