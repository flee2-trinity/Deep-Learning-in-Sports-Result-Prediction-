```python
import csv
filename = '/content/gdrive/My Drive/2021 Fall/Thesis I/Records/PBP1819_0.txt'
with open(filename, newline='') as f:
  reader = csv.reader(f)
  p1content = list(reader)
```

```python
xList = [i for i in range(1, 289)]
y1List = [[]for i in range(300)]
y2List = [[]for i in range(300)]
```

```python
print(len(p1content))
for t in range(len(p1content)):
  s = p1content[t][0].split()
  if len(s) == 4 :
    t = (int(s[1])-1)*72 + (720-int(s[2]))/10
    if int(t)>=0:
      #print(t)
      y1List[int(t)].append(float(s[0]))
      y2List[int(t)].append(float(s[3]))
```

```python
from statistics import mean
y1ListM = list(map(lambda d: mean(d), y1List[:288]))
y2ListM = list(map(lambda d: mean(d), y2List[:288]))
print(y1ListM)
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax_left = plt.subplots()
ax_right = ax_left.twinx()
ax_left.plot(xList, y1ListM, label = "Accuracy", color='green')
ax_right.plot(xList, y2ListM, label = "Diff", color='blue')
plt.legend()
plt.show()
```
