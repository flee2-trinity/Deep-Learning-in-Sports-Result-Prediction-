```python
import csv
filename = '/content/gdrive/My Drive/2021 Fall/Thesis I/Records/elo2_20.txt'
with open(filename, newline='') as f:
  reader = csv.reader(f)
  p1content = list(reader)
```

```python
season = 2019
tpName = 'elo'
```

```python
xList = [i for i in range(163)]
y1List = []
y2List = []
#date, pow, s, season, tp
tempL = list(map(lambda a: a[0].split(), p1content))
cList = list(filter(lambda a: len(a) > 2, tempL))
for x in xList:
  g = list(filter(lambda s: s[4] == tpName and int(s[3]) == season and int(s[0]) == x, cList))[0]
  y1List.append(float(g[1]))
  y2List.append(float(g[2]))
```

```python
from statistics import mean
filename0 = '/content/gdrive/My Drive/2021 Fall/Thesis I/Records/Rank2_22.txt'
with open(filename0, newline='') as f:
  reader = csv.reader(f)
  p0content = list(reader)
y1List0 = [[]for i in range(163)]
y2List0 = [[]for i in range(163)]
for t in range(len(p0content)):
  s = p0content[t][0].split()
  if s[0] != '-' :
    #print(t % 164)
    i = int(s[0])
    y1List0[i].append(float(s[1]))
    y2List0[i].append(float(s[2]))
y1ListM0 = list(map(lambda d: mean(d), y1List0))
y2ListM0 = list(map(lambda d: mean(d), y2List0))
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot(xList, y1List, label = "Power Ranking Elo")
plt.plot(xList, y2List, label = "WinRate Ranking Elo")
plt.plot(xList, y1ListM0, label = "Power Ranking P")
plt.plot(xList, y2ListM0, label = "WinRate Ranking P")
plt.legend()
plt.show()
```
